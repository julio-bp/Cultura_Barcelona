## PROYECTO 3

**Radiografía del consumo cultural en Barcelona (2024): perfiles de consumo, desigualdades territoriales y retos de acceso**

El objetivo de este proyecto es realizar una radiografía del consumo cultural en Barcelona durante el año 2024, analizando los patrones de consumo según variables sociodemográficas (edad, género, nivel educativo, ingresos y situación laboral) y poniéndolos en relación con el contexto urbano y territorial de la ciudad, como la distribución de la población, la vivienda turística y el equipamiento cultural por barrio.

El análisis busca identificar desigualdades en el acceso y la participación cultural, detectar barreras al consumo y explorar posibles desajustes entre la oferta cultural existente y el perfil real de los consumidores culturales, con el fin de aportar información útil para la planificación cultural y urbana.

¿Qué se consume más en Barcelona?

¿Quién consume qué?

¿Por qué no se consume más cultura?

Aplicaciones reales del análisis

1.Diseño de políticas culturales

- Detectar barrios infraequipados.

- Priorizar inversión cultural donde hay población pero poco acceso.

2.Estrategias de democratización cultural

- Programación específica por perfil (edad, estudios, ingresos).

- Reducción de barreras no solo económicas.

3.Planificación urbana

- Evaluar impacto de la vivienda turística sobre el ecosistema cultural.

- Evitar la “cultura para turistas” como único modelo.

4.- Base para análisis longitudinal

Este 2024 es una foto.

- El mismo modelo permite comparar con otros años.
