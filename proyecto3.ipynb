```python
df_poblacion_2024["Valor"] = pd.to_numeric(
    df_poblacion_2024["Valor"],
    errors="coerce"
)
```

```python
df_poblacion_barrio = (
    df_poblacion_2024
    .groupby(
        ["Nom_Districte", "Nom_Barri"],
        as_index=False
    )["Valor"]
    .sum()
)
```

```python
df_poblacion_barrio = df_poblacion_barrio.rename(
    columns={"Nom_Barri": "Barri", "Valor": "habitants"}
)
```

```python
df_poblacion_barrio
```

```python
df_turist
```

```python
turist_2024 = df_turist[
    df_turist["Location type"] == "Barri"
][["Territory", "31 Dec 2024"]].reset_index(drop=True)
```

```python
turist_2024
```

```python
turist_2024["31 Dec 2024"].unique()
```

```python
turist_2024 = turist_2024.rename(columns={
    "Territory": "Barri",
    "31 Dec 2024": "Allotjaments_turistics_2024"
})
```

```python
nivell = []

for valor in turist_2024["Allotjaments_turistics_2024"]:
    if valor > 300:
        nivell.append("Alt")
    elif valor > 100:
        nivell.append("Mitjà")
    else:
        nivell.append("Baix")

turist_2024["nivell_vivienda_turistica"] = nivell
```

```python
turist_2024
```

```python
turist_alt = turist_2024[turist_2024["nivell_vivienda_turistica"] == "Alt"]
turist_mitja = turist_2024[turist_2024["nivell_vivienda_turistica"] == "Mitjà"]
turist_baix = turist_2024[turist_2024["nivell_vivienda_turistica"] == "Baix"]
```

```python
turist_alt
```

```python
turist_baix
```

```python
df_cult_genero = df_cult_genero.loc[:, ~df_cult_genero.columns.str.contains("Unnamed: 1")]
```

```python
columnas_genero = ["Homes", "Dones"]
```

```python
df_cult_genero[columnas_genero] = df_cult_genero[columnas_genero].round(2)
```

```python
df_cult_genero = df_cult_genero.dropna(subset=columnas_genero, how="all")
```

```python
df_cult_genero["genero_mayor_consumo"] = df_cult_genero[columnas_genero].idxmax(axis=1)
df_cult_genero["valor_maximo"] = df_cult_genero[columnas_genero].max(axis=1).round(2)
```

```python
df_cult_genero
```

```python
tabla_hombres_predominan = df_cult_genero[
    df_cult_genero["genero_mayor_consumo"] == "Homes"
][["Pràctiques culturals", "Homes"]].reset_index(drop=True)

tabla_hombres_predominan
```

```python
tabla_mujeres_predominan = df_cult_genero[
    df_cult_genero["genero_mayor_consumo"] == "Dones"
][["Pràctiques culturals", "Dones"]].reset_index(drop=True)

tabla_mujeres_predominan
```
